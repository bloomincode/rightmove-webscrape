```python
from bs4 import BeautifulSoup
import requests

pageres = requests.get('https://www.rightmove.co.uk/property-for-sale/find.html?locationIdentifier=REGION%5E61521&propertyTypes=&includeSSTC=false&mustHave=&dontShow=&furnishTypes=&keywords=')

soup = BeautifulSoup(pageres.content, 'html.parser')
```

```python
soup.title
```

```python
soup.p
```

```python
items = soup.find_all('div', {'class': 'l-searchResult is-list'})
```

```python
for item in items:
    print(item.find('h2', {'class': 'propertyCard-title'}).get_text())
    print(item.find('div', {'class': 'propertyCard-description'}).get_text())
    print(item.find('div', { 'class': 'propertyCard-priceValue'}).get_text())
    print(item.find('img', {'class': 'propertyCard-branchLogo-image'})['alt'])
```
